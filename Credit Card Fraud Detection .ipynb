```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
#importing dataset to a Pandas Dataframe
credit_card_data = pd.read_csv('creditcard.csv')
```

```python
credit_card_data.head()
```

```python
credit_card_data.tail()
```

```python
# Dataset information
credit_card_data.info()
```

```python
# checking the number of missing values in each column
credit_card_data.isnull().sum()
```

```python
# distribution of legit transactions and fraudulent transactions
credit_card_data['Class'].value_counts()
```

# 0 --> normal transactions
# 1 --> fraudulent transactions
# Hence, the data is highly unbalanced

```python
# separating the data for analysis
legit = credit_card_data[credit_card_data.Class == 0]
fraud = credit_card_data[credit_card_data['Class'] == 1]
```

```python
fraud.shape
```

```python
legit.shape
```

```python
#statistical measures of the data
legit['Amount'].describe()
```

```python
fraud['Amount'].describe()
```

```python
# compare the values for both transactions
credit_card_data.groupby('Class').mean()
```


# Under-Sampling of data is observed
# We now build a Sample dataset containing similar distribution of normal transactions and fraudulent transactions, and in order to do this, we take 492 random values from the legit transactions and join it with the fraudulent dataset

```python
legit_sample = legit.sample(n = 492)     #random sampling of legit data
```

```python
# Concatenating two Dataframes
new_dataset = pd.concat([legit_sample, fraud], axis = 0)
```

```python
new_dataset.head()
```

```python
new_dataset.tail()
```

```python
new_dataset['Class'].value_counts()
```

```python
new_dataset.groupby('Class').mean()
```

# Splitting the data into Features and Targets

```python
x = new_dataset.drop(columns = 'Class', axis = 1)
y = new_dataset['Class']
```

```python
print(x)
```

```python
print(y)
```

# Splitting the data into Training data and Test data:

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 2)
```

```python
print(x.shape, x_train.shape, x_test.shape, y_train.shape, y_test.shape)
```

# Machine Learning Model Training
# Logistic Regression

```python
model = LogisticRegression(max_iter = 1000)
```

```python
#training the Logistic Regression Model with Training Data
model.fit(x_train, y_train)
```

# Model Evaluation based on Accuracy Score

```python
# accuracy on training data
x_train_prediction = model.predict(x_train)
training_data_accuracy = accuracy_score(x_train_prediction, y_train)
```

```python
print("Accuracy on Training Data :", training_data_accuracy)
```

```python
# accuracy on training data
x_test_prediction = model.predict(x_test)
test_data_accuracy = accuracy_score(x_test_prediction, y_test)
```

```python
print("Accuracy on Test Data :", test_data_accuracy)
```

# Hence, we see that the Training Data accuracy and Test Data accuracy scores are almost the same, which implies that our Credit Card Fraud Detection model is successful and has an accuracy of around 94%
